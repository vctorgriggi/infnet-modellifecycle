# Carregamento e Exploração Inicial dos Dados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.Series(cancer.target, name="target")
data = pd.concat([X, y], axis=1)
data.head()
```

```python
data.isnull().sum()
```

```python
# 0 = Maligno, 1 = Benigno
sns.countplot(x=y.map({0: 'Maligno', 1: 'Benigno'}))
plt.title("Distribuição das Classes: Tumores Malignos e Benignos")
plt.xlabel("Classe do Tumor")
plt.ylabel("Número de Amostras")
plt.tight_layout()
plt.show()
```

```python
# Estatísticas descritivas das variáveis independentes
X.describe()
```

# Separacão dos Dados e Treinamento do Modelo KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Testa vários valores de K (de 1 a 20) e registra as acurácias
k_values = list(range(1, 21))
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# Plota a acurácia para cada valor de K
plt.figure(figsize=(10, 5))
plt.plot(k_values, accuracies, marker='o')
plt.title("Acurácia para diferentes valores de K")
plt.xlabel("Valor de K")
plt.ylabel("Acurácia")
plt.xticks(k_values)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
melhor_k = k_values[accuracies.index(max(accuracies))]
print(f"Melhor valor de K: {melhor_k} com acurácia de {max(accuracies):.4f}")
```

```python
# Avalia novamente o modelo com o melhor K
knn_final = KNeighborsClassifier(n_neighbors=melhor_k)
knn_final.fit(X_train, y_train)
y_pred_final = knn_final.predict(X_test)

print("\nRelatório de Classificação com Melhor K:\n")
print(classification_report(y_test, y_pred_final, target_names=["Maligno", "Benigno"]))
```

```python
conf_mat = confusion_matrix(y_test, y_pred_final)
sns.heatmap(conf_mat, annot=True, fmt="d", cmap="Blues", xticklabels=["Maligno", "Benigno"], yticklabels=["Maligno", "Benigno"])
plt.title("Matriz de Confusão - KNN com Melhor K")
plt.xlabel("Previsão")
plt.ylabel("Valor Real")
plt.tight_layout()
plt.show()
```

# Geração de Dados Sintéticos com Ruído Gaussiano

```python
# Expande o conjunto de treino com ruído gaussiano
noise_factor = 0.05
X_synthetic = X_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=X_train.shape)
y_synthetic = y_train.copy()

# Concatena original + sintético
X_augmented = pd.concat([X_train, pd.DataFrame(X_synthetic, columns=X.columns)], axis=0)
y_augmented = pd.concat([y_train, y_synthetic], axis=0)

# Treina KNN com dados aumentados
knn_augmented = KNeighborsClassifier(n_neighbors=melhor_k)
knn_augmented.fit(X_augmented, y_augmented)
y_pred_aug = knn_augmented.predict(X_test)

# Avaliação com dados sintéticos
acc_aug = accuracy_score(y_test, y_pred_aug)
print(f"\nAcurácia com dados sintéticos: {acc_aug:.4f} (antes: {max(accuracies):.4f})")
```

# Diferença entre Classificação e Regressão

```python
from IPython.display import Markdown as md

md("""
### Classificação vs Regressão

- **Classificação** é utilizada quando o objetivo é prever **categorias ou classes** (ex: benigno vs maligno).
- **Regressão** é usada para prever **valores numéricos contínuos**, como área, volume ou tempo.

Neste trabalho, utilizamos:
- Utilizamos classificação para prever o tipo do tumor, no qual pode ser 0 ou 1.
- Usaremos regressão para estimar a **área média do tumor** com base em atributos numéricos.
""")
```

# Regressão Linear para Estimar a Área Média do Tumor

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import skew, kurtosis
```

```python
# Define a variável alvo da regressão
y_reg = X["mean area"]
X_reg = X.drop(columns=["mean area"])

Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

reg = LinearRegression()
reg.fit(Xr_train, yr_train)
```

```python
# Faz previsões
y_pred_reg = reg.predict(Xr_test)

# Avaliação do modelo
r2 = r2_score(yr_test, y_pred_reg)
print(f"R² do modelo de regressão: {r2:.4f}")
```

```python
# Gráfico Y real vs Y estimado
plt.figure(figsize=(6, 5))
plt.scatter(yr_test, y_pred_reg, alpha=0.7)
plt.title("Y Real vs Y Estimado")
plt.xlabel("Y Real")
plt.ylabel("Y Estimado")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Gráfico Y real vs Resíduo Normalizado
residuals = yr_test - y_pred_reg
plt.figure(figsize=(6, 5))
plt.scatter(yr_test, residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.title("Y Real vs Resíduo")
plt.xlabel("Y Real")
plt.ylabel("Resíduo")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histograma dos resíduos
plt.figure(figsize=(6, 4))
sns.histplot(residuals, kde=True)
plt.title("Distribuição dos Resíduos")
plt.xlabel("Resíduo")
plt.tight_layout()
plt.show()
```

```python
print("\nEstatísticas dos resíduos:")
print(f"Média: {residuals.mean():.4f}")
print(f"Variância: {residuals.var():.4f}")
print(f"Assimetria: {skew(residuals):.4f}")
print(f"Curtose: {kurtosis(residuals):.4f}")
```
